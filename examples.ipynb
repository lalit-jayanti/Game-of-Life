```python
import numpy as np
from game_of_life import gameOfLife
```

## Examples

### B3/S23 Life

```python
gol = gameOfLife(height=100,
                 width=100,
                 birth_rules=[3],
                 survival_rules=[2, 3])

for _ in range(50):
    gol.step()
    
gol.save('results/life.gif', steps=150, cmap='gray')
```

### B1357/S1357 Replicator

```python
gol = gameOfLife(height=100,
                 width=100,
                 birth_rules=[1, 3, 5, 7],
                 survival_rules=[1, 3, 5, 7])

board = np.zeros((200,200))
board[99:102, 99:102] = np.array([[1, 1, 1],
                                  [1, 0, 1],
                                  [1, 1, 1]])

gol.reset(board)

gol.save('results/replicator.gif', steps=50, cmap='Blues', fps=8)
```
